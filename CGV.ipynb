```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup

## CGV의 크롤링하기 알맞지 않는 사이트인것을 인지했다 

class CGV_movie_select(object):
    def __init__(self, chorme_driver):
        self.url = "https://www.cgv.co.kr"
        self.driver = webdriver.Chrome(chorme_driver)
    
    def search(self, movie_name):
        self.driver.get(self.url)
        self.driver.implicitly_wait(2)  

        input_element =  self.driver.find_element_by_xpath('//*[@id="header_keyword"]')
        

        if input_element:
            # input 요소에 값을 입력
            input_element.send_keys(movie_name)
    
            # 입력 후 Enter 키 입력
            input_element.send_keys(Keys.ENTER)
        else:
            print("Input 요소를 찾을 수 없습니다.")
            
        html = self.driver.page_source
        
        soup = BeautifulSoup(html)
        
        movie_page = soup.select_one(".searchingMovieResult_list")
        
        movie_page_link = self.url+movie_page.a.attrs['href']
        
        self.driver.get(movie_page_link)
        
        html = self.driver.page_source
        
        
        return html
    
    def extract(element ,html):
        result = []
        soup = BeautifulSoup(html)
        for item in element:
            result.append(soup.select_one(item))
        
        return result
    
    
    
    def driver_close(self):
        self.driver.close()
        
                
        

        


element = []
element.append(".box-contents>")

driver = r"C:\Users\HAMA\Desktop\crawling\chromedriver-win64\chromedriver.exe"


test = CGV_movie_select(driver)
html = test.search("파묘")
test.driver_close()

```
